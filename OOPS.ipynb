# Encapsulation

```python
class Employee:

    def __init__(self):
        self.__salary = 50000

    def get_salary(self):
        return self.__salary

emp = Employee()

print(emp.get_salary())
```

```python
class Car:
    def __init__(self, car_name, car_model, car_color):
        self.__car_name = car_name
        self.__car_model = car_model
        self._car_color = "Black"

    def get_details(self):
        print(f"Car name is {self.__car_name} and car model is {self.__car_model}")

car = Car("Tata", "Tiago")
car.get_details()
```

# Inheritance

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def sound(self):
        print(f"This {self.name} animal sound")

class Dog(Animal):
    def bark(self):
        print("Bark")


obj = Dog("Dog")
obj.sound()
obj.bark()
```

# Polymorphism

```python
# same method, different behavior because different object so different input and output 
# Overriding method:- 
class Animal:
    def sound(self):
        print("Animal sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

d = Dog()
d.sound()
```

# Abstraction

```python
# Hiding implementation details and showing only essential functionality.
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car started")

c = Car()
c.start()
```

# Method Types

```python
# Instance Method 
class Demo:
    def display(self):
        print("Instance")
d = Demo()
print(d)
```

```python
d.display()
```

```python
# Class Method 

class Demo:
    @classmethod
    def display(cls):
        print(" This is a class method")
d = Demo()
d.display()
```

```python
# Static Method

class Sum:
    @staticmethod
    def add(a, b):
        return a+b
        
Sum().add(4, 6)        
```

```python
# Composition
class Engine:

    def start(self):
        print("Engine Started")

class Car:

    def __init__(self):
        self.engine = Engine()

c = Car()

c.engine.start()
```

```python
class Vehicle:
    def start(self):
        print("Vehicle is started")
class Bike:
    def __init__(self):
        self.vehicle = Vehicle()

c = Bike()
c.vehicle.start()
```

# Create a Bank Account class

```python
class BankAccount:

    def __init__(self, name, balance=0):
        self.name = name
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient Balance")

    def show_balance(self):
        print("Balance:", self.__balance)


account = BankAccount("Sonam", 1000)

account.deposit(500)
account.withdraw(200)
account.show_balance()
```
